#### Замена типов данных 

«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайтов. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута. Округлим секунды до минут, так как округляется каждый звонок. 

Мегабайты до гигабайтов округлим в следующей главе, после группировки интернет-сессий по месяцам, так как до гигабайтов округляется сумма использованных мегабайтов за месяц, а не каждая интернет-сессия.


#### Округление секунд до минут

Видно, что в датасете минута делится на 100 частей, а не на 60.   
Округлим значения до ближайших больших целых чисел (минут) и переведем float в int.

Для округления числа до ближайшего большего целого, используем функцию np.ceil.   
Заменим тип данных на int, так как числа должны быть целыми:

```python
data_calls.head(5)
```

```python
data_calls['duration'] = data_calls['duration'].apply(pd.to_numeric, errors='coerce')
data_calls['duration'] = np.ceil(data_calls['duration'])
data_calls['duration'] = data_calls['duration'].astype('int')
```

Видно, что тип данных изменился на int, а значения округлились до ближайших целых:

####  Расчет помесячной выручки с каждого абонента

Рассчитаем помесячную выручку с каждого абонента.
Вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножим на значение из тарифного плана. 

Для этого используем три функции. Создадим три новых столбца - с доходом отдельно по звонкам, сообщениям и интернету - 'income_call', 'income_message', 'income_internet'.  

Минуты и сообщения сверх тарифа стоят 1 руб. по тарифу 'Ультра' и 3 руб.  по тарифу 'Смарт'.  
1Гб интернета сверх тарифа стоит 150 руб. по тарифу 'Ультра' и 200 руб.  по тарифу 'Смарт'.  

В конце создадим новый столбец 'total_income', где будет суммирована общая прибыль, включая абонентскую плату.


```python
def income_call(row):
    if row['tariff'] == 'ultra':
        if row['calls_min'] >  row['minutes_included']:
            return (row['calls_min'] - row['minutes_included']) * row['rub_per_minute']
            
    if row['tariff'] == 'smart':
        if row['calls_min'] >  row['minutes_included']:
            return (row['calls_min'] - row['minutes_included']) * row['rub_per_minute']
            
            
merged['income_call'] = merged.apply (lambda row: income_call(row), axis=1)
```

```python
def income_message(row):
    
    if row['tariff'] == 'ultra':
        if row['message_quantity'] >  row['messages_included']:
            return  (row['message_quantity'] - row['messages_included']) * row['rub_per_message']
            
    if row['tariff'] == 'smart':
        if row['message_quantity'] >  row['messages_included']:
            return  (row['message_quantity'] -  row['messages_included']) * row['rub_per_message']
            
            
merged['income_messages'] = merged.apply (lambda row: income_message(row), axis=1)
```

```python
def income_internet(row):
    if row['tariff'] == 'ultra':
        if row['internet_traffic'] >  row['gb_per_month_included']:
            return  (row['internet_traffic'] - row['gb_per_month_included']) * row['rub_per_gb'] 
    
    if row['tariff'] == 'smart':
        if row['internet_traffic'] >  row['gb_per_month_included']:
            return  (row['internet_traffic'] - row['gb_per_month_included']) * row['rub_per_gb'] 
            
            
merged['income_internet'] = merged.apply (lambda row: income_internet(row), axis=1)
```

Затем прибавим абонентскую плату и создадим столбец 'total income'. В нем можно увидеть общий доход по каждому клиенту помесячно:

```python
merged['total_income'] = merged[['income_call', 'income_messages', 'income_internet','rub_monthly_fee']].sum(axis=1)
```

### Проверка гипотез

#### Гипотеза: средняя выручка от пользователей тарифов «Ультра» и «Смарт» различаются.

Проверим гипотезу: средняя выручка от пользователей тарифов «Ультра» и «Смарт» различается.

Чтобы подтвердить это, нужно отвергнуть гипотезу H₀:  

H₀ - средняя выручка от пользователей тарифов «Ультра» и «Смарт» равна    
H₁ - средняя выручка от пользователей тарифов «Ультра» и «Смарт» различается

Нам интересны различия в обе стороны - меньшую и большую, поэтому гипотеза H₁ двусторонняя.

H₀ всегда формулируется так, чтобы использовать знак равенства.  
H₁ - альтернативная гипотеза, которая противоречит нулевой гипотезе, — либо нулевая гипотеза не отвергается, либо отвергается и принимается H₁.

Возьмем стандартный уровень критической значимости alpha - 0.05.

Используем ttest для двух независимых совокупностей 'ind':

```python
ultra_clients_income = merged.query('tariff == "ultra"')
smart_clients_income = merged.query('tariff == "smart"')
```

```python
ultra_clients_income = ultra_clients_income['total_income']
smart_clients_income = smart_clients_income['total_income']  
print('cреднее Ультра: ',pd.Series(ultra_clients_income).mean())
print('cреднее Смарт: ',pd.Series(smart_clients_income).mean())
alpha = .05 
results = st.ttest_ind(
    ultra_clients_income,
    smart_clients_income)

print('p-значение:', results.pvalue) 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Мы выбрали значения alpha .05, так как это общепринятый критический уровень статистической значимости. Но видно, что p-value настолькое маленькое, что даже при аlpha .01 нулевая гипотеза была бы отвергнута. 

У нас появилось основание отбросить нулевую гипотезу, так как при условии, что нулевая гипотеза верна, существует слишком маленькая вероятность получить данные, не противоречащие ей.

#### Гипотеза: cредняя выручка от пользователей из Москвы отличается от выручки от пользователей из других регионов.

Проверим гипотезу: cредняя выручка от пользователей из Москвы отличается от выручки от пользователей из других регионов различается.

Чтобы подтвердить это, нужно отвергнуть гипотезу H₀:  

H₀ - cредняя выручка от пользователей из Москвы равна выручке от пользователей из других регионов.  
H₁ - cредняя выручка от пользователей из Москвы не равна выручке от пользователей из других регионов.

Нам интересны различия в обе стороны - меньшую и большую, поэтому гипотеза H₁ двусторонняя.

H₀ всегда формулируется так, чтобы использовать знак равенства.

H₁ - альтернативная гипотеза, которая противоречит нулевой гипотезе, — либо нулевая гипотеза не отвергается, либо отвергается и принимается H₁.

Возьмем стандартный уровень критической значимости alpha - 0.05.

Используем ttest для двух независимых совокупностей 'ind':

```python
clients_income_moscow = merged.query('city == "Москва"')
clients_income_not_moscow = merged.query('city != "Москва"')
```

```python
clients_income_moscow = clients_income_moscow['total_income']
clients_income_not_moscow = clients_income_not_moscow['total_income']  
print('cреднее Москва:',pd.Series(clients_income_moscow ).mean())
print('cреднее не Москва:',pd.Series(clients_income_not_moscow).mean())
alpha = .05 
results = st.ttest_ind(
    clients_income_moscow,
    clients_income_not_moscow)

print('p-значение:', results.pvalue) 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Мы получили p-value 0.54, то есть при условии, что нулевая гипотеза верна, вероятность получить данные, не противоречащие ей, составляет 0.54 (или 54%). Это значение выше  уровня критической значимости alpha 0.05, а значит, нет оснований отбрасывать нулевую гипотезу.
